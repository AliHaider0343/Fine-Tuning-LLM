# Loading Base Model and Tokenizer

```python
# Model from Hugging Face hub
base_model = "NousResearch/CodeLlama-7b-hf"

# New instruction dataset
pinescript_dataset = "AliHaider0343/Indicators-Dataset"

# Fine-tuned model
new_model = "code-llama-2-7b-5e-0.1-pinescript-indicators"
```

```python
compute_dtype = getattr(torch, "float16")

quant_config = BitsAndBytesConfig(
    load_in_4bit=True,
    bnb_4bit_quant_type="nf4",
    bnb_4bit_compute_dtype=compute_dtype,
    bnb_4bit_use_double_quant=False,
)
```

```python
model = AutoModelForCausalLM.from_pretrained(
    base_model,
    quantization_config=quant_config,
    device_map={"": 0}
)
model.config.use_cache = False
model.config.pretraining_tp = 1
```

```python
tokenizer = AutoTokenizer.from_pretrained(base_model)
tokenizer.add_eos_token = True
tokenizer.pad_token_id = 0
tokenizer.padding_side = "right"
```

# Loading Formated Data Set

```python
from datasets import load_dataset

dataset = load_dataset(pinescript_dataset, split="train")
```

```python
dataset
```

# Lora & Model Hyperparameter Configurations

```python
peft_params = LoraConfig(
    lora_alpha=32,
    lora_dropout=0.1,
    r=16,
    bias="none",
    task_type="CAUSAL_LM",
)
```

```python

training_params = TrainingArguments(
    output_dir="/content/drive/MyDrive/finetuned_llama_models1/results_code_llama-5e-0.1",
    num_train_epochs=2,
    per_device_train_batch_size=2,
    gradient_accumulation_steps=2,
    optim="paged_adamw_32bit",
    save_steps=50,
    logging_steps=50,
    learning_rate=2e-4,
    weight_decay=0.001,
    fp16=False,
    bf16=False,
    max_grad_norm=0.3,
    max_steps=-1,
    warmup_ratio=0.03,
    group_by_length=True,
    lr_scheduler_type="linear",
    report_to="tensorboard"
)
```

```python
trainer = SFTTrainer(
    model=model,
    train_dataset=dataset,
    peft_config=peft_params,
    dataset_text_field="text",
    max_seq_length=None,
    tokenizer=tokenizer,
    args=training_params,
    packing=False,
)
```

# Model Trainings

```python
trainer.train()
```

# Model saving locally and Generated Inference on HF

```python
trainer.model.save_pretrained(new_model)
trainer.tokenizer.save_pretrained(new_model)
```

```python
trainer.push_to_hub("engrzulqarnain/"+new_model)
```
